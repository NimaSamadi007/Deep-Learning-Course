# Problem 2, World Cup Matches Predictions

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch

%matplotlib inline
```

## A) Loading data

```python
raw_df = pd.read_csv("./data/HW2_data/international_matches.csv")
raw_df.tail(10)
```

## B) Plotting data

```python
home_team_fifa_rank = raw_df.home_team_fifa_rank.to_numpy()
away_team_fifa_rank = raw_df.away_team_fifa_rank.to_numpy()
match_result = raw_df.home_team_result.to_numpy()

fig, ax = plt.subplots(figsize=(12, 8))
cdict = {'Win': 'red', 'Draw': 'black', 'Lose': 'blue'}
for res in ['Win', 'Draw', 'Lose']:
    idx = np.where(match_result == res)
    ax.scatter(home_team_fifa_rank[idx], away_team_fifa_rank[idx], c=cdict[res], label=res)
ax.legend(loc='upper right')
ax.grid(True)
ax.set_title("Match results")
ax.set_xlabel("Home team FIFA ranking")
ax.set_ylabel("Away team FIFA ranking")
plt.show()
```
